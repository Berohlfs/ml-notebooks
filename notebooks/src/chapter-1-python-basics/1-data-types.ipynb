# Data Types


```python
# IMMUTABLE

gravity = 0 # Int

weight = 34.5 # Float

is_valid = True # Bool

name = "John" # Str

last_name = None # None

tuple_of_names = ("John", "Jane", "Jim", "Jill") # Tuple

print("Int:", gravity)
print("Float:", weight)
print("Bool:", is_valid)
print("Str:", name)
print("None:", last_name)
print("Tuple:", tuple_of_names)

# MUTABLE

list_of_names = ["John", "Jane", "Jim", "Jill"] # List

dict_of_names = {"John": 1, "Jane": 2, "Jim": 3, "Jill": 4} # Dict

set_of_names = {"John", "Jane", "Jim", "Jill"} # Set

print("List:", list_of_names)
print("Dict:", dict_of_names)
print("Set:", set_of_names)
```

## Operations

## Int

```python
a, b = 10, 3

# Arithmetic operations
print("Arithmetic:")
print(f"{a} + {b} = {a + b}")  # Addition
print(f"{a} - {b} = {a - b}")  # Subtraction
print(f"{a} * {b} = {a * b}")  # Multiplication
print(f"{a} / {b} = {a / b}")  # Division (returns float)
print(f"{a} // {b} = {a // b}")  # Floor division
print(f"{a} % {b} = {a % b}")  # Modulo
print(f"{a} ** {b} = {a ** b}")  # Exponentiation

# Comparisons
print("\nComparisons:")
print(f"{a} == {b}: {a == b}")
print(f"{a} != {b}: {a != b}")
print(f"{a} > {b}: {a > b}")
print(f"{a} < {b}: {a < b}")

# Type conversion
print("\nType conversion:")
print(f"int('42'): {int('42')}")
print(f"int(3.7): {int(3.7)}")  # Truncates

# Useful methods
print(f"\nabs(-5): {abs(-5)}")
print(f"max(1, 5, 3): {max(1, 5, 3)}")
print(f"min(1, 5, 3): {min(1, 5, 3)}")
print(f"sum([1, 2, 3]): {sum([1, 2, 3])}")
```

## Float

```python
x, y = 3.14, 2.5

print("Float Operations:")
print(f"{x} + {y} = {x + y}")
print(f"{x} * {y} = {x * y}")
print(f"round({x}): {round(x)}")
print(f"round({x}, 1): {round(x, 1)}")

# Precision handling
print(f"\nPrecision:")
print(f"0.1 + 0.2: {0.1 + 0.2}")  # Floating point precision issue
print(f"round(0.1 + 0.2, 1): {round(0.1 + 0.2, 1)}")

# Type conversion
print(f"\nfloat('3.14'): {float('3.14')}")
print(f"float(5): {float(5)}")
```

## Bool

```python
# Logical operations
p, q = True, False
print(f"p = {p}, q = {q}")
print(f"p and q: {p and q}")
print(f"p or q: {p or q}")
print(f"not p: {not p}")

# Truthiness (Python's truthy/falsy values)
print("\nTruthiness:")
print(f"bool(1): {bool(1)}")
print(f"bool(0): {bool(0)}")
print(f"bool(''): {bool('')}")  # Empty string is False
print(f"bool('text'): {bool('text')}")  # Non-empty is True
print(f"bool([]): {bool([])}")  # Empty list is False
print(f"bool([1]): {bool([1])}")  # Non-empty is True
print(f"bool(None): {bool(None)}")  # None is False

# Short-circuit evaluation
print("\nShort-circuit:")
result = True or print("This won't print")
result = False and print("This won't print either")
```

## Str

```python
text = "Hello World"
name = "Python"

print("String Operations:")

# Concatenation and repetition
print(f"'{text}' + ' ' + '{name}': {text + ' ' + name}")
print(f"'{name}' * 3: {name * 3}")

# Indexing and slicing
print(f"\nIndexing & Slicing:")
print(f"text[0]: {text[0]}")
print(f"text[-1]: {text[-1]}")  # Last character
print(f"text[0:5]: {text[0:5]}")  # Slicing
print(f"text[:5]: {text[:5]}")  # From start
print(f"text[6:]: {text[6:]}")  # To end

# Step examples (positive step)
print(f"\nStep Examples (Positive Step):")
print(f"text[::2]: {text[::2]}")  # Every 2nd character from start
print(f"text[::3]: {text[::3]}")  # Every 3rd character from start
print(f"text[1::2]: {text[1::2]}")  # Every 2nd character starting from index 1
print(f"text[0:5:2]: {text[0:5:2]}")  # Every 2nd character from 0 to 5
print(f"text[2:8:2]: {text[2:8:2]}")  # Every 2nd character from 2 to 8

# Step examples (negative step)
print(f"\nStep Examples (Negative Step):")
print(f"text[::-1]: {text[::-1]}")  # Reverse entire string
print(f"text[::-2]: {text[::-2]}")  # Every 2nd character in reverse
print(f"text[-1::-1]: {text[-1::-1]}")  # Reverse from last character
print(f"text[5:0:-1]: {text[5:0:-1]}")  # Reverse from index 5 to 0 (exclusive)
print(f"text[8:2:-2]: {text[8:2:-2]}")  # Every 2nd character in reverse from 8 to 2

# Common string methods
print(f"\nString Methods:")
print(f"text.upper(): {text.upper()}")
print(f"text.lower(): {text.lower()}")
print(f"text.title(): {text.title()}")
print(f"text.replace('World', 'Python'): {text.replace('World', 'Python')}")
print(f"text.split(): {text.split()}")
print(f"'  hello  '.strip(): {'  hello  '.strip()}")
print(f"text.startswith('Hello'): {text.startswith('Hello')}")
print(f"text.endswith('World'): {text.endswith('World')}")
print(f"text.find('World'): {text.find('World')}")
print(f"'hello' in text.lower(): {'hello' in text.lower()}")

# String formatting (f-strings - Python 3.6+)
age = 25
print(f"\nF-string formatting:")
print(f"My name is {name} and I'm {age} years old")
print(f"Formatted number: {3.14159:.2f}")
```

## None

```python
value = None
print(f"value is None: {value is None}")  # Use 'is' not '=='
print(f"value is not None: {value is not None}")

# Common pattern: default value
result = value or "Default Value"
print(f"None or 'Default': {result}")

# Checking for None before operations
data = None
if data is not None:
    print("Data exists")
else:
    print("Data is None")
```

## Tuple

```python
tup = (1, 2, 3, 4, 5)

print("Tuple Operations:")

# Indexing and slicing (same as lists)
print(f"tup[0]: {tup[0]}")
print(f"tup[-1]: {tup[-1]}")
print(f"tup[1:3]: {tup[1:3]}")

# Unpacking
a, b, c = (1, 2, 3)
print(f"\nUnpacking: a={a}, b={b}, c={c}")

# Extended unpacking
first, *rest = tup
print(f"first, *rest = {tup}: first={first}, rest={rest}")

# Tuple operations
print(f"\nlen(tup): {len(tup)}")
print(f"3 in tup: {3 in tup}")
print(f"tup + (6, 7): {tup + (6, 7)}")  # Concatenation
print(f"tup * 2: {tup * 2}")  # Repetition
print(f"tup.count(3): {tup.count(3)}")
print(f"tup.index(3): {tup.index(3)}")

# Immutability demonstration
print("\nImmutability:")
print("Tuples are immutable - cannot modify after creation")
print("tup[0] = 10  # This would raise TypeError")
```

## List

```python
lst = [1, 2, 3]

print("List Operations:")

# Adding elements
lst.append(4)
print(f"After append(4): {lst}")

lst.extend([5, 6])
print(f"After extend([5, 6]): {lst}")

lst.insert(0, 0)
print(f"After insert(0, 0): {lst}")

# Removing elements
lst.remove(0)  # Remove first occurrence
print(f"After remove(0): {lst}")

popped = lst.pop()  # Remove and return last element
print(f"After pop(): {lst}, popped={popped}")

popped = lst.pop(0)  # Remove and return element at index
print(f"After pop(0): {lst}, popped={popped}")

# List operations
print(f"\nlen(lst): {len(lst)}")
print(f"2 in lst: {2 in lst}")
print(f"lst.index(3): {lst.index(3)}")
print(f"lst.count(2): {lst.count(2)}")

# Sorting
lst2 = [3, 1, 4, 1, 5]
lst2.sort()
print(f"After sort(): {lst2}")

lst3 = [3, 1, 4, 1, 5]
sorted_lst = sorted(lst3)  # Returns new list
print(f"sorted([3,1,4,1,5]): {sorted_lst}")

# Reversing
lst.reverse()
print(f"After reverse(): {lst}")

# Slicing (creates new list)
print(f"lst[1:3]: {lst[1:3]}")
print(f"lst[::-1]: {lst[::-1]}")  # Reverse copy

# List comprehension (powerful!)
squares = [x**2 for x in range(5)]
print(f"\nList comprehension [x**2 for x in range(5)]: {squares}")

evens = [x for x in range(10) if x % 2 == 0]
print(f"Evens [x for x in range(10) if x % 2 == 0]: {evens}")
```

## Dict

```python
d = {"name": "Alice", "age": 30, "city": "NYC"}

print("Dict Operations:")

# Accessing values
print(f"d['name']: {d['name']}")
print(f"d.get('age'): {d.get('age')}")
print(f"d.get('country', 'Unknown'): {d.get('country', 'Unknown')}")  # Default value

# Adding/Updating
d["country"] = "USA"
print(f"After d['country'] = 'USA': {d}")

d.update({"age": 31, "job": "Engineer"})
print(f"After update(): {d}")

# Removing
removed = d.pop("job")
print(f"After pop('job'): {d}, removed={removed}")

# Dict operations
print(f"\nlen(d): {len(d)}")
print(f"'name' in d: {'name' in d}")
print(f"d.keys(): {list(d.keys())}")
print(f"d.values(): {list(d.values())}")
print(f"d.items(): {list(d.items())}")

# Iterating
print("\nIterating:")
for key, value in d.items():
    print(f"  {key}: {value}")

# Dict comprehension
squares_dict = {x: x**2 for x in range(5)}
print(f"\nDict comprehension {{x: x**2 for x in range(5)}}: {squares_dict}")

# Merging dictionaries (Python 3.9+)
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}
merged = {**d1, **d2}
print(f"Merged dicts: {merged}")
```

## Set

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

print("Set Operations:")

# Adding/Removing
set1.add(5)
print(f"After add(5): {set1}")

set1.remove(5)  # Raises KeyError if not found
print(f"After remove(5): {set1}")

set1.discard(10)  # Safe remove (no error if not found)
print(f"After discard(10): {set1}")

popped = set1.pop()  # Remove and return arbitrary element
print(f"After pop(): {set1}, popped={popped}")

# Set operations
print(f"\nset1: {set1}, set2: {set2}")
print(f"set1 | set2 (union): {set1 | set2}")
print(f"set1 & set2 (intersection): {set1 & set2}")
print(f"set1 - set2 (difference): {set1 - set2}")
print(f"set1 ^ set2 (symmetric difference): {set1 ^ set2}")

# Set methods
print(f"\nset1.union(set2): {set1.union(set2)}")
print(f"set1.intersection(set2): {set1.intersection(set2)}")
print(f"set1.difference(set2): {set1.difference(set2)}")
print(f"set1.issubset(set2): {set1.issubset(set2)}")
print(f"set1.issuperset(set2): {set1.issuperset(set2)}")

# Set operations
print(f"\nlen(set1): {len(set1)}")
print(f"3 in set1: {3 in set1}")

# Set comprehension
evens_set = {x for x in range(10) if x % 2 == 0}
print(f"Set comprehension: {evens_set}")

# Removing duplicates from list
lst_with_dupes = [1, 2, 2, 3, 3, 3]
unique = list(set(lst_with_dupes))
print(f"\nRemove duplicates: {lst_with_dupes} -> {unique}")
```
